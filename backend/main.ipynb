```python
def outlier_removal(df):
    processed_df = df.copy()

    for col in processed_df.select_dtypes(include=['number']).columns:
        Q1 = processed_df[col].quantile(0.25)
        Q3 = processed_df[col].quantile(0.75)
        IQR = Q3 - Q1
        if IQR > 0:
         lower_bound = Q1 - 3 * IQR
         upper_bound = Q3 + 3 * IQR
         print(f"Before Clipping ({col}):")
         print(processed_df[col].describe())
         processed_df[col] = np.clip(processed_df[col] , lower_bound , upper_bound)
         print(f"After Clipping ({col}):")
         print(processed_df[col].describe())  # Print after clipping
         print("="*50)
    
    return processed_df
```

```python
def skewness_handling(df):
    processed_df = df.copy()
    for col in processed_df.select_dtypes(include=['number']).columns:
            skewness = processed_df[col].skew()
            if skewness > 1:
                if(processed_df[col] > 0).all():
                    processed_df[col] = np.log1p(processed_df[col])
                else:
                    processed_df[col] = np.sqrt(processed_df[col] - processed_df[col].min() + 1)
            elif -0.5 < skewness < 0.5:
                pass
            elif skewness < -1:
                processed_df[col] = np.sqrt(processed_df[col])
    return processed_df
        
```

```python
# When you define bins in pd.cut(), each bin includes the values in the left boundary but excludes the values in the right boundary by default (except for the last bin).

# bins = [-0.1, 0, 10000, 20000, 30000, 40000, 50000, 60000]
# (-0.1, 0] â†’ Includes 0, but nothing less than 0

# (0, 10000] â†’ Includes values >0 and â‰¤ 10000

# (10000, 20000] â†’ Includes values >10000 and â‰¤ 20000, and so on...

# ðŸ“Œ Why -0.1 is Used?
# If you defined bins as [0, 10000, 20000, ...], then:

# Values exactly equal to 0 would not be included in any bin!

# To avoid this, -0.1 is added so that 0 falls into the first bin.

# ðŸ“Œ Why Was There No -0.1 in Your Dataset?
# Even though -0.1 is not present in the dataset, it is only used as a left boundary for binning. It does not appear in the final resultsâ€”it just ensures 0 gets categorized properly.
```
